1. To what does a relative path refer?

A relative path referes to the location that is relative to the current directory.

A relative path  is relative to the program’s current working directory

Relative path is defined as the path related to the present working directly(pwd). 

It starts at your current directory and never starts with a \ .

Eg:

..\

3. What do the functions os.getcwd() and os.chdir() do?

```python
'''
os.getcwd()
-> The method returns a string which represents the location of current work directory.
Example:
'''
import os
cwd=os.getcwd()
print('Current Working Directory : '+cwd)
```

```python
'''
os.chdir()
-> The method is used to change the curent working directory(CWD) to specified path.
-> It doesn’t return any value.
Example:
'''
import os
print('Directory Before Change : '+os.getcwd())
os.chdir(r'C:\Users')
print('!!! Directory Changed !!')
print('Directory After Change : '+os.getcwd())
```

6. What are the three “mode” arguments that can be passed to the open() function?

```python
Three “mode” arguments that can be passed to the open() function are:
a. 'r' -  To read file
b. 'a' -  To append file
c. 'w' -  To write in file
```

```python
'''Example'''
f=open('file1.txt','w')
f.write('Hello! You have written it!')
f.close()
f1=open('file1.txt','a')
f1.write('\nFile appended!')
f1.close()
f2=open('file1.txt','r')
print(f2.read())
f2.close()
```

7. What happens if an existing file is opened in write mode?

If existing file is opened in write mode, it will overwrites the existing file from the scratch.

```python
'''Example'''
f=open('file1.txt','r')
print('Org file content : \n'+f.read())
f.close()
f1=open('file1.txt','w')
f1.write('This is new content!')
f1.close()
f2=open('file1.txt','r')
print('\nAfter write mode : \n'+f2.read())
f.close()
```

9. What data structure does a shelf value resemble?

Shelf value resemble to dictionary data structure.
Binary shelf files is saved using the shelve module.

```python
'''Example'''
import shelve
sf=shelve.open('mydata')
city=['Delhi','Mumbai']
sf['city']=city
sf.close()
```
