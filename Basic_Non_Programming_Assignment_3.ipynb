5. How many global scopes are there in a Python program? How many local scopes?

```python
'''

There is one global scope, and a local scope is created whenever a function is called.

The scope defines the accessibility of the python object. 
The particular coding region where variables are visible is known as scope.

The namespace is the unique identification of the variable or the method. 
It tells the python interpreter about the name of the object and the location from where it is trying to access it.

The Namespaces are searched for scope resolution according to the LEGB rule.
The LEGB stands for :L: Local, E: Enclosed, G: Global, B: Built-in.
The sequence of LEGB is important. 
The variable is first searched in Local, followed by Enclosed, then global and finally built-in.


Global Variable in Python – Variables created outside the function are called global variables. Anyone can access the global variable, both inside and outside of a function. 

Parameters and variables that are assigned in a called function are said to exist in that function’s local scope.
A variable that exists in a local scope is called a local variable.


1. Local Variables Cannot Be Used in the Global Scope
Eg:
def func():
    var = 31337
func()
print(var)  -> var is local variable, it can't be used in global scope.

2. Local Scopes Cannot Use Variables in Other Local Scopes
Eg:
def spam():
    eggs = 99
    bacon()
    print(eggs)

def bacon():
    ham = 101
    eggs = 0

spam()

Result --> 99

3. Global Variables Can Be Read from a Local Scope
Eg:
def spam():
    print(eggs)
eggs = 42     -----> Golbal Variable
spam()
print(eggs)

4. Local and Global Variables with the Same Name
def spam():
    eggs = 'spam local'
    print(eggs)    --> prints 'spam local'

eggs = 'global'
spam()
print(eggs)        --> prints 'global'


There are four rules to tell whether a variable is in a local scope or global scope:

1. If a variable is being used in the global scope (that is, outside of all functions), then it is always a global variable.
2. If there is a global statement for that variable in a function, it is a global variable.
3. Otherwise, if the variable is used in an assignment statement in the function, it is a local variable.
4. But if the variable is not used in an assignment statement, it is a global variable.


'''
```

13. What can you do to save a programme from crashing if it encounters an error?

```python
'''
To save programme from crashing if it encounters an error, 
we can place the line of code that might cause an error in a try clause (Exception Handling).
'''

try:
    print(10/0)
except Exception as e:
    print('ERROR : '+str(e))
```
