1. What are escape characters, and how do you use them?

```python
'''
An escape character lets use characters that are otherwise impossible to put into a string.
An escape character consists of a backslash (\) followed by the character you want to add to the string.
There are several different different escape characters.
1. \'      Prints Single Quote
2. \"      Prints Double Quote
3. \t      Prints Tab
4. \n      Prints newline(line break)
5. \r      It will just work as cursor has been shifted to the beginning of the string or line.
6. \\      Prints \
'''
```

9. What are the methods for right-justifying, left-justifying, and centering a string?

```python
'''
1. right-justifying
-> rjust()
-> string method that return a padded version of the string they are called on, with spaces inserted to justify the text.
Eg:
'Dell'.rjust(10,'*')
'******Dell'

2. left-justifying
-> ljust()
-> string method that return a padded version of the string they are called on, with spaces inserted to justify the text.
Eg:
'Dell'.ljust(10,'-')
'Dell------'

3. centering
-> center()
-> string method that return a padded version of the string and centers the text(if any).
Eg:
'Dell'.center(10,'-')
'---Dell---'
'''
```
