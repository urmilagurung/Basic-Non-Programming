4. From a Match item, how do you get the actual strings that match the pattern?

The group() method returns strings of the matched text.

Passing 0 or nothing to the group() method will return the entire matched text.

Adding parentheses will create groups in the regex.

Eg:

```python
idnumber= re.compile(r'(\w\w\w)(\d\d\d)(\w\d\w)(\d\d\d\w)')
idnumber=idnumber.search('Employee ID : DLD121W0V679D')
print('Group 4: '+str(idnumber.group(4)))
```

5. In the regex which created from the r'(\d\d\d)-(\d\d\d-\d\d\d\d)', what does group zero cover? Group 2? Group 1?

Here,

Group 2 - (\d\d\d-\d\d\d\d) 

-> The second set of parentheses in a regex string will be group 2.

Group 1 - (\d\d\d)

-> The first set of parentheses in a regex string will be group 1.

Eg:

```python
phone = re.compile(r'(\d\d\d)-(\d\d\d-\d\d\d\d)')
p=phone.search('My number is 901-890-2301.')
print('Group 1 : '+str(p.group(1)))
print('Group 2 : '+str(p.group(2)))
```

7. The findall() method returns a string list or a list of string tuples. What causes it to return one of the two options?

The findall() method will return the strings of every match in the searched string.

Eg:

```python
'''
1. If the regex has no groups, a list of strings is returned. 
'''
import re
postal_code=re.compile(r'\d\d\d\d\d\d')
p=postal_code.findall('Delhi : 111022, Mumbai:451001')
print('List of Postal Code found: '+str(p))    
```

```python
'''
2. If the regex has groups, a list of tuples of strings is returned.
'''
import re
postal_code=re.compile(r'(\d\d\d\d)-(\d\d)')
p=postal_code.findall('Delhi : 1110-22, Mumbai:4510-01')
print('List of tuples of Postal Code found: '+str(p))    
```

8. In standard expressions, what does the | character mean?

The | character is called a pipe. 

The | character signifies matching "either, or" between two groups.

It can be used to match one of many expressions.

Eg:

The regular expression r'Batman|Robin' will match either 'Batman' or 'Robin'.

```python
dc=re.compile(r'Batman|Robin')
dc=dc.search('Batman and Robin')
dc.group()
```

14. What is the difference between .*? and .*?

```python
'''
.*   try to match as much text as possible. The dot-star uses greedy mode.
'''

greedy = re.compile(r'<.*>')
g = greedy.search('<To serve man> for dinner.>')
g.group()
```

```python
'''
.*?  To match nongreedy way
'''

nongreedy = re.compile(r'<.*?>')
ng = nongreedy.search('<To serve man> for dinner.>')
ng.group()
```

16. What is the procedure for making a normal expression in regax case insensitive?

The procedure for making a normal expression in regax case insensitive is by passing below as a second argument to re.compile().
1. re.IGNORECASE
2. re.I
