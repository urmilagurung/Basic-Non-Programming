1. How do you distinguish between shutil.copy() and shutil.copytree()?

```python
''' 
shutil.copy () method is used to copy the content of source file to destination file or directory. 
This method returns a string which represents the path of newly created file.
'''
import os,shutil

path=r'C:\Users\CHRISTINA GURUNG\Course\Assignment\Basic_Non_Programming\Test1'
print('Before copying file : ',os.listdir(path))

source=r'C:\Users\CHRISTINA GURUNG\Course\Assignment\Basic_Non_Programming\file1.txt'

permission=os.stat(source).st_mode
print('File permission mode : ',permission,'\n')

destination = r'C:\Users\CHRISTINA GURUNG\Course\Assignment\Basic_Non_Programming\Test1\file1(copy).txt'

d=shutil.copy(source,destination)

print('After copying file : ',os.listdir(path))

permission=os.stat(destination).st_mode
print('File permission mode : ',permission)

print("Destination path:", d)
```

```python
'''
shutil.copytree() method recursively copies an entire directory tree rooted at source (src) to the destination directory.
The destination directory, named by (dst) must not already exist. 
This method returns a string which represents the path of newly created directory.
'''

import os,shutil

path=r'C:\Users\CHRISTINA GURUNG\Course\Assignment\Basic_Non_Programming\Test1'
print('Before copying file : ',os.listdir(path))

source=r'C:\Users\CHRISTINA GURUNG\Course\Assignment\Basic_Non_Programming\Test1\Test2'

destination = r'C:\Users\CHRISTINA GURUNG\Course\Assignment\Basic_Non_Programming\Test1\Test3'

d=shutil.copytree(source,destination)

print('After copying file : ',os.listdir(path))
print("Destination path:", d)
```

5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). Copy these files from whatever location they are in to a new folder.

```python
import os,shutil

from pathlib import Path
p=Path.cwd()

def getfile(start,filter_exe=None):
    for root, directories, files in os.walk(start):
        for file in files:
            if file.upper().endswith(filter_exe):
                yield os.path.join(root,file)

def filtercopy(source,destination,filter_exe=None):
    for file in getfile(source,filter_exe):
        shutil.copy(file,destination)
        print('The file has been copied : ',file)

        
filtercopy(p/"oldjpg",p/"newjpg",".JPG")        
```
