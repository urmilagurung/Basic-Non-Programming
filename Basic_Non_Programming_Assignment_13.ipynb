2.What do you pass to csv.reader() and csv.writer() to create reader and writer objects?

You pass a File object, obtained from a call to open().

A Writer object lets you write data to a CSV file. To create a Writer object, you use the csv.writer() function. 
First, call open() and pass it 'w' to open a file in write mode will create the object you can then pass to csv.writer()  to create a Writer object.

To read a CSV file with the csv module, first open it using the open() function, just as you would any other text file.
csv.reader() function will return a Reader object for you to use.

3. What modes do File objects for reader and writer objects need to be opened in?

File objects need to be opened in read-binary ('rb') for reader objects and write-binary ('wb') for writer objects.
