1. Is the Python Standard Library included with PyInputPlus?

PyInputPlus is not a part of the Python Standard Library.
It needs to be downloaded separately using Pip.

PyInputPlus contains functions similar to input() for several kinds of data: numbers, dates, email addresses, and more. 

If the user ever enters invalid input, such as a badly formatted date or a number that is outside of an intended range, PyInputPlus will reprompt them for input just like our code in the previous section did. PyInputPlus also has other useful features like a limit for the number of times it reprompts users and a timeout if users are required to respond within a time limit.

```python
'''Example'''
import pyinputplus as pypi
pypi.inputEmail('Enter Email ID:')
```

2. Why is PyInputPlus commonly imported with import pyinputplus as pypi?

The as pyip code in the import statement saves from typing pyinputplus each time we want to call a PyInputPlus function. Therefore, shorter name pyip is used. 

3. How do you distinguish between inputInt() and inputFloat()?

```python
'''
inputInt() accepts integers.
Eg:
'''
import pyinputplus as pyip
pyip.inputInt('Enter a number greater than 0 but less than 10 :',min=0,max=10)
```

```python
'''
inputFloat() accepts a Float number.
Eg:
'''
import pyinputplus as pyip
pyip.inputFloat('Enter a number greater than 1.5 but less than 8.75 :',min=1.5,max=8.75)
```

4. Using PyInputPlus, how do you ensure that the user enters a whole number between 0 and 99?

```python
from pyinputplus import inputInt
inputInt('Enter a number greater than 0 but less than 99 :',greaterThan=0,lessThan=100)
```

5. What is transferred to the keyword arguments allowRegexes and blockRegexes?

```python
'''
allowRegexes allow expressions passed along with user input.
Example:
'''

from pyinputplus import inputInt
inputInt('Enter a number greater than 0 but less than 99 :',greaterThan=0,lessThan=100,allowRegexes=[r'(I|V|X|L|C)+',r'one'])
```

```python
'''
blockRegexes doesn't allow expressions passed along with user input.
Example:
'''

from pyinputplus import inputInt
inputInt('Enter a number greater than 0 but less than 99 :',greaterThan=0,lessThan=100,blockRegexes=[r'(I|V|X|L|C)+',r'one'])
```

6. If a blank input is entered three times, what does inputStr(limit=3) do?

If blank input is entered 3 times then limit arguments will cause the function to raise a RetryLimitException.

```python
import pyinputplus as pyin
try:
    pyin.inputStr('Enter your name :',limit=3)
except pyin.RetryLimitException:
    print('RetryLimitException')
```

7. If blank input is entered three times, what does inputStr(limit=3, default='hello') do?

If blank input is entered 3 times then function returns default value instead of raising exception.

```python
import pyinputplus as pyin
bill=pyin.inputStr('Enter Name : ',limit=3, default='hello')
print(f'{bill}! This is Bill!')
```
