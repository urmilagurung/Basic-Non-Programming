3. What is the most significant distinction between a dictionary and a list?

```python
'''
Most significant distinction is items stored in list are ordered whereas in Dictionary, its unordered.Because dictionaries are not ordered, they can’t be sliced like lists.

List:
List is a collection of index values
List is created by placing elements in [ ] seperated by commas “, “
The elements are accessed via indices.
The indices of list are integers starting from 0.
The order of the elements entered are maintained.
Eg:
[10,'Hello',(10,20,30),100]

Dictionary:
Dictionary is a hashed structure of key and value pairs.
Dictionary is created by placing elements in { } as “key”:”value”, each key value pair is seperated by commas “, ”
The elements are accessed via key-values.
The keys of dictionary can be of any data type.
There is no guarantee for maintaining order.
Eg:
{'a':'apple,'b':'ball'}

It is more efficient to use a dictionary for lookup of elements because it takes less time to traverse in the dictionary than a list.
'''
```

8. How do you "pretty print" dictionary values using which module and function?

```python
#"pretty print" is helpful when you want a cleaner display of the items in a dictionary than what print() provides. 

from pprint import pprint
message = 'This is month of May'
count = {}

for character in message:
    count.setdefault(character, 0)
    count[character] = count[character] + 1

pprint(count)
```
