 3. Describe three different data types.

```python
'''
There different data types are as below:

1. Numeric
--> It represents data which has numeric value.
--> Numeric value can be of 3 types. They are as below:
    i. Integers
    --> This value is represnted by integer class.
    --> It contains positive or negative whole numbers(without fraction/decimal).
    --> Eg: -20; 910 
    ii. Float
    --> It is represented by float class.
    --> It is real number with floating point representation.
    --> It is specified by decimal point.
    --> Optionally, the character e or E followed by a positive or negative integer may be appended to specify scientific           notation.
    --> Eg: -23.90; 80.99
    iii. Complex Numbers
    --> Complex number is represented by complex class. 
    --> It is specified as (real part) + (imaginary part)j. 
    --> Eg: – 2+3j

2. List
--> It is ordered collection of data.
--> It is very flexible as item in the list need not be of same type.
--> It is enclosed in square brackets.
--> It supports indexing and slicing.
--> It is mutable. Therefore can be reassigned.
--> Eg: [20,'Hi',['Billy',('Dell')]]

3. Dictionary
--> It is unordered collection of data values, used to store data values like map(which unlike other Data Types that hold       only single value as an element).
--> It holds data in key:value pair.
--> Each key-value pair in a Dictionary is separated by a colon :, whereas each key is separated by a ‘comma’.
--> Dictionary can be created by placing a sequence of elements within curly {} braces, separated by ‘comma’.
--> Values in a dictionary can be of any datatype and can be duplicated, whereas keys can’t be repeated and must be
    immutable.
--> Dictionary can also be created by the built-in function dict()
--> Eg: {1: 'Geeks', 2: 'For', 3: 'Geeks'}

'''
```
