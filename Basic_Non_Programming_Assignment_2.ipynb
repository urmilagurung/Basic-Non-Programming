2. What are the three different types of Boolean operators?

```python
'''
Boolean operators form the basis of mathematical sets and database logic.
They connect search words together to either narrow or broaden your set of results.
The three basic boolean operators are: AND, OR, and NOT.
    
i. AND
--> It narrow results.
--> It tells the database that ALL search terms must be present in the resulting records.
--> Eg: if a and b in c

ii. OR
--> It connect two or more similar concepts (synonyms)
--> It broaden results, telling the database that ANY of your search terms can be present in the resulting records.
--> Eg : if a or b in c

iii. NOT
--> It exclude words from search.
--> It narrow search, telling the database to ignore concepts that may be implied by your search terms.
--> Eg: if not a==10
'''
```
